### Вводные слова о Pandas <a class="anchor" id="introduction"></a>

**Pandas** - это библиотека, содержащая высокоуровневые структуры данных и инструменты для анализа данных. Так себе определение, если честно... **Pandas** - это библиотека для работы с данными, в частности - это в основном работа с табличными данными. С помощью этой библиотеки вы можете осуществлять наверное практически (по-любому что-то нельзя, но я не в курсе) любые операции с табличными данными, включая их простую визуализацию (хотя для полноценной визуализации лучше использовать другие библиотеки).

**Pandas** является надстройкой над библиотекой **NumPy**, именно поэтому важно знать основную структуру и операции, доступные в **NumPy**.

В **Pandas** имеются две основные структуры данных - это объекты **Series** (для одномерного случая, считайте отдельные колонки) и DataFrame (для двухмерного случая, счтайте таблицы). Вообще имеется еще объект **Panel** (для трехмерного случая), а еще многомерные объекты, но это просто для вашей эрудици =) 

Импортирование библиотеки осуществляется с помощью давно принятого в обществе префикса **pd** (не раскачивайте лодку, используя другие префиксы!!!)

```python
import pandas as pd
```

Рассмотрим в качестве введения объекты класса **Series**, которые являются одномерными массивами индексированных данных. На самом деле можно считать, что **Series** - это сплав **NumPy** массива (**ndarray**) и питоновского словаря (**dict**). Почему? Потому что они ведут себя, как NumPy массивы, однако индексируются не только неявно с помощью индексации от нуля (первый элемент) и дальше, а индексируются с помощью произвольных ключей ОДНОГО типа.

При создании объекта **Series** можно не указывать индексы, тогда они представляют из себя привычные для массивов целые числа, начинающиеся с нуля (первый элемент).

```python
ser = pd.Series([1, 2, 3, 4, 5])
ser
```

Как видно, данный объект имеет индексы, значения и тип значений. Чтобы получить отдельно значения или индексы нужно воспользоваться соответстенно атрибутами **values** и **index**.

```python
print(ser.values)
print(type(ser.values))
```

Полученный массив значений является уже непосредственно NumPy массивом, с которыми вы уже встречались.

В свою очередь атрибут **index** возвращает специальный numpy-подобный объект еще одного основополагающего класса **Index**.

```python
ser.index
```

При создании объекта **Series** можно вручную указывать список индексов.

```python
ser = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
ser
```

```python
ser.index
```

В случае индексов работает такая же логика, как и с ключами словарей. К примеру, ключом может являться неизменяемый кортеж.

```python
ser = pd.Series([1, 2, 3, 4, 5], index=[('a', 1), ('b', 2), ('c', 3), ('d', 4), ('e', 5)])
ser
```

```python
ser.index
```

Логика обращения к элементам **Series** и взятия срезов довольно запутанная. К элементам можно обращаться, как к элементам словарям или же как к элементам массива. Однако, иногда эта логика абсолютно непрозрачная. В дальнейшем мы рассмотрим методы обращения к элементам на примере таблиц, которые не будут запутывать. Сейчас здесь будет продемонстрированы примеры неочевидного обращения к элементам **Series**. **НЕ ЗАПОМИНАЙТЕ ИХ**, просто убедитесь, что лучше так не делать.

```python
ser
```

```python
print(ser[0])
print(ser[-1])
print(ser[1:3])
```

```python
print(ser[('d', 4)])
```

Также есть возможность использовать смешенную логику массива и словаря в плане взятия срезов.

```python
print(ser[('a', 1):('d', 4)])
```

<div class="alert alert-block alert-danger">
<b>АЛАРМ:</b> Обратите внимание, что здесь последний элемент включается, в отличии от срезов по индексам обычно списка/массива.
</div>

А что если сделать такие индексы.

```python
ser = pd.Series([1, 2, 3, 4, 5], index=[1, 2, 3, 4, 5])
ser
```

Обращаться неявно к элементам массива через индексацию, начинающуюся с нуля уже нельзя

```python
ser[0]
```

Надеюсь вы убедились, что в таком виде запомнить индексацию без последующих путанниц, довольно сложно. Теперь можете забыть все это, как страшный сон.

Напоследок рассмотрим еще один способ создания объекта **Series**. Как уже было сказано, данный объект - это смесь массива и словаря. Поэтому создавать объекты данного типа можно и с помощью обычного словаря.

```python
d = {'a': 10, 'b': 20, 'c': 30}

ser = pd.Series(d)
ser
```

С помощью списков словарей также можно создать объект класса **DataFrame** (двухмерный массив, читай табличка). Такой способ создания может пригодиться для быстрого создания тестовой таблички для проверки какой-либо операции над ней.

```python
d = {'a': 10, 'b': 20, 'c': 30}
e = {'a': 40, 'b': 50, 'c': 60}

df = pd.DataFrame([d, e], index=['ind_1', 'ind_2'])
df
```
