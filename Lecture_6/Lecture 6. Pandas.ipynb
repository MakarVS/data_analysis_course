### Вводные слова о Pandas <a class="anchor" id="introduction"></a>

**Pandas** - это библиотека, содержащая высокоуровневые структуры данных и инструменты для анализа данных. Так себе определение, если честно... **Pandas** - это библиотека для работы с данными, в частности - это в основном работа с табличными данными. С помощью этой библиотеки вы можете осуществлять наверное практически (по-любому что-то нельзя, но я не в курсе) любые операции с табличными данными, включая их простую визуализацию (хотя для полноценной визуализации лучше использовать другие библиотеки).

**Pandas** является надстройкой над библиотекой **NumPy**, именно поэтому важно знать основную структуру и операции, доступные в **NumPy**.

В **Pandas** имеются две основные структуры данных - это объекты **Series** (для одномерного случая, считайте отдельные колонки) и DataFrame (для двухмерного случая, счтайте таблицы). Вообще имеется еще объект **Panel** (для трехмерного случая), а еще многомерные объекты, но это просто для вашей эрудици =) 

Импортирование библиотеки осуществляется с помощью давно принятого в обществе префикса **pd** (не раскачивайте лодку, используя другие префиксы!!!)

```python
import pandas as pd
```

Рассмотрим в качестве введения объекты класса **Series**, которые являются одномерными массивами индексированных данных. На самом деле можно считать, что **Series** - это сплав **NumPy** массива (**ndarray**) и питоновского словаря (**dict**). Почему? Потому что они ведут себя, как NumPy массивы, однако индексируются не только неявно с помощью индексации от нуля (первый элемент) и дальше, а индексируются с помощью произвольных ключей ОДНОГО типа.

При создании объекта **Series** можно не указывать индексы, тогда они представляют из себя привычные для массивов целые числа, начинающиеся с нуля (первый элемент).

```python
ser = pd.Series([1, 2, 3, 4, 5])
ser
```

Как видно, данный объект имеет индексы, значения и тип значений. Чтобы получить отдельно значения или индексы нужно воспользоваться соответстенно атрибутами **values** и **index**.

```python
print(ser.values)
print(type(ser.values))
```

Полученный массив значений является уже непосредственно NumPy массивом, с которыми вы уже встречались.

В свою очередь атрибут **index** возвращает специальный numpy-подобный объект еще одного основополагающего класса **Index**.

```python
ser.index
```

При создании объекта **Series** можно вручную указывать список индексов.

```python
ser = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
ser
```

```python
ser.index
```

В случае индексов работает такая же логика, как и с ключами словарей. К примеру, ключом может являться неизменяемый кортеж.

```python
ser = pd.Series([1, 2, 3, 4, 5], index=[('a', 1), ('b', 2), ('c', 3), ('d', 4), ('e', 5)])
ser
```

```python
ser.index
```

Логика обращения к элементам **Series** и взятия срезов довольно запутанная. К элементам можно обращаться, как к элементам словарям или же как к элементам массива. Однако, иногда эта логика абсолютно непрозрачная. В дальнейшем мы рассмотрим методы обращения к элементам на примере таблиц, которые не будут запутывать. Сейчас здесь будет продемонстрированы примеры неочевидного обращения к элементам **Series**. **НЕ ЗАПОМИНАЙТЕ ИХ**, просто убедитесь, что лучше так не делать.

```python
ser
```

```python
print(ser[0])
print(ser[-1])
print(ser[1:3])
```

```python
print(ser[('d', 4)])
```

Также есть возможность использовать смешенную логику массива и словаря в плане взятия срезов.

```python
print(ser[('a', 1):('d', 4)])
```

<div class="alert alert-block alert-danger">
<b>АЛАРМ:</b> Обратите внимание, что здесь последний элемент включается, в отличии от срезов по индексам обычно списка/массива.
</div>

А что если сделать такие индексы.

```python
ser = pd.Series([1, 2, 3, 4, 5], index=[1, 2, 3, 4, 5])
ser
```

Обращаться неявно к элементам массива через индексацию, начинающуюся с нуля уже нельзя

```python
ser[0]
```

Надеюсь вы убедились, что в таком виде запомнить индексацию без последующих путанниц, довольно сложно. Теперь можете забыть все это, как страшный сон.

Напоследок рассмотрим еще один способ создания объекта **Series**. Как уже было сказано, данный объект - это смесь массива и словаря. Поэтому создавать объекты данного типа можно и с помощью обычного словаря.

```python
d = {'a': 10, 'b': 20, 'c': 30}

ser = pd.Series(d)
ser
```

С помощью списков словарей также можно создать объект класса **DataFrame** (двухмерный массив, читай табличка). Такой способ создания может пригодиться для быстрого создания тестовой таблички для проверки какой-либо операции над ней.

```python
d = {'a': 10, 'b': 20, 'c': 30}
e = {'a': 40, 'b': 50, 'c': 60}

df = pd.DataFrame([d, e], index=['ind_1', 'ind_2'])
df
```

- sep - разделитель, по-умолчанию пытается распознать разделитель (данный аргумент исключетелен для csv файлов);

Остальные доп. аргументы могу использоваться в любом другом методе **read**

- dtype - словарь, с помощью которого явно задаются типы столбцов таблицы;
- na_values - заполнение пропусков конкретным значением.

Для этого вставьте строку ниже без знака решетки в терминал (либо замените знак решетки восклицательным знаком, что будет означать работу в терминале и уберите лишний пробел).

```python
# pip install pyxlsb
```

```python
%%time

from pyxlsb import open_workbook as open_xlsb

df = []

with open_xlsb('Materials_sixth_lecture/MDM.xlsb') as wb:
    with wb.get_sheet(1) as sheet:
        for row in sheet.rows():
            df.append([item.v for item in row])
            
data_xlsb = pd.DataFrame(df[1:], columns=df[0])       
```

```python
data_xlsb.head()
```

Обработка отсутствующих значений

При анализе предыдущих параметров видно, что отсутствующие значения присутствуют только в столбце **price**, а также в столбце **price_round_hundred**, так как он образован от первого.

Метод **isnull()** генерирует булеву маску на место пропусков, проставляя True.

```python
data[['price', 'price_round_hundred']].isnull()
```

Существует противоположность метода **isnull()** - **notnull()** 

```python
data[['price', 'price_round_hundred']].notnull()
```

Полученные булевы маски можно использовать в качестве фильтров, однако чаще используются встроенные методы для обработки пустых значений:
- **dropna()** - удаление пустых значений;
- **fillna()** - заполнение пустых значених.

По-умолчанию метод **dropna** удаляет все строки, содержащие пустые значения.

```python
print(len(data))
print(len(data.dropna()))
```

Можно удалять столбцы, в которых есть пустые значения

```python
data.dropna(axis=1).head()
```

Однако, при таком подходе может удалиться много "хороших данных".

Можно удалять данные с помощью изменения аргумента **how**. По умолчанию **how = 'any'**, т.е. как раз удаляет строку или столбец при хотя бы одном пустом значении. Передав аргументу значени **how='all'** будут удаляться только те строки/столбцы в которых все данные пропущены.

Или же можно задать порог минимального кол-ва НЕпропущенных значений, до достижения которого строка/столбец не удаляется.

```python
data.dropna(axis=0, thresh=9)
```

Для заполнения пропущенных данных как уже говорилось используется метод **fillna()**, в который передается значение, заменяющее пропуски.

```python
data = data.fillna(data['price'].median())
data.info()
```

Обращение к элементам таблицы

Существует несколько способов обращения к элементам таблицы. Первый из них это поочередное обращение к столбцу и потом к элементу из него по индексу.

```python
data['company'][0]
```

Второй способ, это использование атрибута **loc**, в который через **КВАДРАТНЫЕ** скобки передается индекс и имя столбца через запятую.

```python
data.loc[0, 'company']
```

Третий способ - это использование метода **iloc**, в который также через **КВАДРАТНЫЕ** скобки передаются номера индексов массива так, как будто это NumPy массив (то есть индексация является целочисленной и начинается с нуля).

```python
data.iloc[0, 0]
```

Два вышеприведенных атрибута удобно использовать для вычленения части таблицы.

```python
data.loc[3:10, ['company', 'price']]
```

### Объединение и пересечение таблиц <a class="anchor" id="join"></a>

В библиотеке Pandas имеется метод аналогичный методу **np.concatenate()** в NumPy - **pd.concat()**. Он работает аналогично с одним лишь отличием - по-умолчанию индексы и столбцы при объединении сохраняются.

```python
data_1 = pd.DataFrame({'A':['A0', 'A1'], 'B': ['B0', 'B1']})
data_1
```

```python
data_2 = pd.DataFrame({'A':['A2', 'A3'], 'B': ['B2', 'B3']})
data_2
```

```python
pd.concat([data_1, data_2], axis=0)
```

```python
pd.concat([data_1, data_2], axis=1)
```

Проблему дублирования индексов можно решать двумя способами:
- не позволять объединение при наличии дублей (с помощью атрибута **verify_integrity**, который по-умолчанию **False**);
- игнорировать индексы (с помощью атрибута **ignore_index**, который по-умолчанию **False**).

```python
pd.concat([data_1, data_2], axis=0, verify_integrity=True)
```

```python
pd.concat([data_1, data_2], axis=0, ignore_index=True)
```

```python
pd.concat([data_1, data_2], axis=1, verify_integrity=True)
```

В случае колонок

```python
pd.concat([data_1, data_2], axis=1, ignore_index=True)
```

### Групировка и агрегирование таблиц <a class="anchor" id="groupby"></a>

```python
data.head()
```

Как уже говорилось выше, общие агрегированные статистики можно посмотреть с помощью метода **describe()**. Однако, можно также взять отдельный столбец и использовать агрегированные методы аналогичные методам NumPy библиотеки.

```python
print(data['company'].count())
print(data['price'].mean())
```

Как и в SQL в Pandas реализована операция группировки - **GroupBy**

Агрегация

```python
data.groupby('company').aggregate([min, max])
```

```python
data.groupby('company').aggregate({'price': ['min', 'max', 'count'],
                                   'horsepower': ['mean', 'std']})
```

Фильтрация

```python
data.groupby('company').filter(lambda x: x['price'].count() > 2)
```
