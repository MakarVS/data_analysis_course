## Условные операторы <a class="anchor" id="if"></a>

Составная инструкция **if**, включающая инструкции **if**, **elif** и **else**, обеспечивает условное выполнение блоков кода.

Вначале всегда выполняется инструкция **if**, если выражение в нем дает истинное значение, то выполняются инструкции, находящиеся внутри, и на этом выполнение всей условной конструкции заканчивается. Если выражение ложное, то поочередно выполняются проверки всех выражений **elif** до первого истинного значения, после этого выполняются вложенные инструкции и выполнение всей условной конструкции заканчивается. При условии, что все выражения ложны - выполняется инструкция **else**.

<div class="alert alert-block alert-success">
<b>Справка:</b> Интсрукции elif и else являются необязательными!
</div>

```
if выражение:   
    инструкции   
elif выражение:   
    инструкции   
elif выражение:    
    инструкции    
...   
elif выражение:    
    инструкции   
else:    
    инструкции
```

<div class="alert alert-block alert-danger">
<b>АЛАРМ:</b> Заметьте, что в Python блоки кода отделяются пробелами (никаких фигурных скобок). Данная особенность языка делает код довольно хорошо читаемым. Вложенный блок кода должен отделяться от внешнего четырьмя пробелами (четыре пробела являются лишь рекомендуемыми по стилю программирования PEP 8). По-умолчанию, юпитер и другие среды разработки автоматически сдвигают каретку, когда понимают что дальше находится вложенный блок. Однако, если вы удалите данные пробелы - Python сгенерирует исключение!
</div>

*Пример*.    
Если текущая прибыль profit_now больше прибыли в предыдущем месяце profit_last - вывести на экран фразу "Как видите, благодаря нашей интенсивной работе, прибыль растет опережающими темпами".    
Если текущая прибыль profit_now меньше прибыли в предыдущем месяце profit_last, но больше нуля - вывести на экран фразу "Из-за случайных колебаний рынка наша прибыль в текущем месте (profit_now) меньше, чем в предыдущем месяце (profit_last). Однако мы все равно в плюсе!!!".   
В остальных случаях выводить фразу - "Ситуация на рынке была катастрофическая, мы ничего не могли поделать. Посмотрите на конкуретна X у него дела обстоят еще хуже!!!".      
В выводимых фразах вместо profit_now и proft_last выведите точные цифры, используя форматируемые строки.

````python
profit_now = 10_000_000
profit_last = 7_000_000
````

## Цикл while <a class="anchor" id="while"></a>

Цикл while обеспечивает многократное выполнение инструкций, образующих тело цикла, пока выражение после while истинное. Процесс выполнения инструкций внутри цикла будет продолжаться, пока выражение не станет ложным.

```
while выражение:
    инструкции
```

<div class="alert alert-block alert-danger">
<b>АЛАРМ:</b> Запомните, что код и выражение в цикле должны быть написаны таким образом, чтобы условие в выражении в конечном счете приобрело ложный результат. Иначе цикл никогда не сможет выполниться. Исключением использования бесконечного цикла считается использование иструкций <a href="#break">break</a> и <a href="#continue">continue</a> (будет рассказываться чуть позже).
</div>

*Пример*.    
Ваш начальник каждый раз спрашивает вас: "Ты сделал ту задачу, о которой я просил?".    
Пока у начальника счетчик, меряющий его раздражительность, не перевалил определенную границу на его вопрос о выполненных задачах выводим на экран пронумерованную по порядку фразу - "1. Да уже почти готово, осталось пару правок внести".     
Изначальный счетчик раздражительности начальства равен нулю, при каждой такой фразе счетчик увеличивается на один. При достижении предела раздражительности (10), выводим на экран фразу - "Если через 10 минут работа не будет сделана, можешь зарабатывать себе на жизнь только преподаванием курсов по питону!"

## Цикл for <a class="anchor" id="for"></a> 

Цикл for обеспеcпечивает многократное выполнение блока инструкций,управляемых итерируемым выражением.

```
for текущий_элемент in итерируемое_выражение:
    инструкции
```

Что такое итерируемое_выражение?

Грубо говоря итерируемое выражение - это такой объект Python по элементам которого мы можем пройтись. Любые последовательности являются итерируемыми объектами - строки, кортежи, списки. Другие коллекции, такие как словари, множества - тоже являются итерируемыми объектами.

<div class="alert alert-block alert-success">
<b>Справка:</b> Кому интересно глубоко понять, что такое итерируемый объект - почитайте про методы iter( ) и next( )
</div>

Посмотрим на примерах

````python
word = 'Hello'

for letter in word:
    print(letter)
````

По сути в переменную letter на каждом шаге начинает поочередно записываться элемент последовательности word. На первой итерации в переменную letter записывается следующая буква "H", выполняется блок итераций и т.д.

Аналогично с кортежами или списками

````python
list_of_names = ['Ivan', 'Petr', 'Vitaliy']

for name in list_of_names:
    print(f'Hello, {name}!')
````

В случае словарей в элементы будут записываться ключи

````python
dict_of_ages = {'Ivan': 29, 'Petr': 33, 'Vitaliy': 24}

for name in dict_of_ages:
    print(name)
````

Хотите обратиться к ключам? Можете сделать так

````python
dict_of_ages['Ivan']
````

````python
for name in dict_of_ages:
    print(f'Возраст {name} - {dict_of_ages[name]}')
````

Кроме всего прочего можно указывать несколько элементов как при присваивании с распаковкой.

````python
dict_of_ages
````

Напомню, что метод items возвращает список (не совсем список на самом деле, а специальный итерируемый объект, но это оффтоп =) ) кортежей из ключа и значения по этому ключу.

````python
dict_of_ages.items()
````

````python
a, b = (10 , 20)
print(a)
print(b)
````

````python
for name, age in dict_of_ages.items():
    print(f'Возраст {name} - {age}')
````

<div class="alert alert-block alert-danger">
<b>АЛАРМ:</b> Запомните, НИКОГДА не изменяйте в теле цикла итерируемое выражение!
</div>

*Пример*. Создадим список фамилий и проитерируемся по нему. На каждой итерации мы будем выводить фамилию и добавлять в конец списка ту же фамилию. В итоге мы получим бесконечный цикл, так как итерируемое выражение (в нашем случае список) на каждом шаге увеличивается на один.

````python
list_of_surnames = ['Ivanov', 'Petrov', 'Medvedev']

for surname in list_of_surnames:
    print(surname)
    list_of_surnames.append(surname)
````

### Метод range <a class='anchor' id='range'></a>

Давайте посмотрим на более привычный вид цикла, как например в C++ - это обход последовательности целых чисел.

````python
for i in range(5):
    print(i)
````

Метод range имеет следующий вид

$$range(начало,конец,шаг)$$

где начало конец и шаг работают также, как в индексах. Начало (не обязательно) - по-умолчанию всегда ноль, шаг (не обязателен) - по-умолчанию равен одному, конец (обязателен) - не входит в последовательность.

````python
for i in range(10, 3, -1):
    print(i)
````

Метод range на самом деле возвращает итерируемый объект из последовательности цифр

````python
range(5)
````

Превратим его в список

````python
list(range(5))
````

### Генераторы <a class='anchor' id='generator'></a> 

Восхитительной особенностью языка Python являются генераторы

<img src="Image_second_lecture/generators.jpg" alt="Генераторы" width="400">

Генераторы позволяют на лету с помощью цикла for создавать коллекции - списки, кортежи, множества, словари. 

Самым распространеным является генератор списков (list comprehension). Генератор имеет следующий вид:

```
[выражение for элемент in итерируемый_объект дополнительное_предложение]
```
Дополнительное предложение может быть двух видов: еще одним циклом (для создания вложенных циклов) и условной конструкцией if:
```
for элемент in итерируемый_объект
ИЛИ
if выражение
```

Генераторы списков являются синтаксической плюшкой. Например код

````python
result = [i for i in range(5)]
result
````

Эквивалентен коду

````python
result_1 = []

for i in range(5):
    result_1.append(i)
    
result_1
````

А код

````python
result_2 = [i for i in range(10) if i % 2 == 0]
result_2
````

Эквивалентен

````python
result_3 = []

for i in range(10):
    if i % 2 == 0:
        result_3.append(i)
        
result_3
````

Посмотрим на создание списка с помощью вложенных циклов

````python
result_4 = [i*j for i in range(1, 4) for j in range(4, 6)]
result_4
````

Можно записать как

````python
result_5 = []

for i in range(1, 4):
    for j in range(4, 6):
        result_5.append(i*j)
        
result_5
````

Генератора кортежей не существует, вы можете увидеть, что такая запись возвращает сам объект генератора

````python
(i for i in range(3))
````

Но можно сделать так

````python
tuple(i for i in range(3))
````

Генераторы множеств (set comprehension) работают аналогично спискам, только в фигурных скобках

````python
results_6 = {i**3 for i in range(4)}
results_6
````

Генераторы словарей (dict comprehension) используют такой же синтаксис, как у множеств, однако перед for должно быть выражение $ключ:значение$

````python
results_7 = {f'{i}^3': i**3 for i in range(4)}
results_7
````

## Инструкция break <a class='anchor' id='break'></a>

Инструкция break разрешается использовать ТОЛЬКО в теле цикла. Выполнение данной инструкции приводит к преждевременному выходу из цикла.

<div class="alert alert-block alert-success">
<b>Справка:</b> Обычно инструкции break и continue записываются в условные конструкции if и поэтому выполняются по условию
</div>

````python
count = 0 

while True:
    count += 1
    if count > 10:
        break

print(f'Счетчик остановился на {count}')
````

````python
for i in range(20):
    if i > 5:
        break
    else:
        print(i)
````

<div class="alert alert-block alert-success">
<b>Справка:</b> Если вы помните Python в таких выражениях воспринимает любой объект, как объект логического выражения
</div>

````python
count = 10

while count:
    count -= 1
    print(count)
    break

print(f'Счетчик остановился на {count}')
````

<div class="alert alert-block alert-success">
<b>Справка:</b> Не записываете в выражения условных операторов или циклов условия типа - x == True, x is True или bool(x). Это плохой тон в программировании на Python.
</div>

## Инструкция continue <a class='anchor' id='continue'></a>

Инструкция continue, как и break может встречаться лишь в теле цикла. Выполнение данной инструкции прерывает выполнение текущей операции  и передает управление в начало цикла.

````python
for i in range(2, 8):
    if i > 3 and i < 6:
        continue
        print('jahsdljhaskdjhashd')
    else:
        print(i)
````

````python
for i in range(5):
    print(i)
````
