### Общие принципы
* Содержание блокнота - это компьютерная программа, то есть инструкция для компьютера, которая позволяет ему прийти к нужному вам результату
* Вы можете создавать и выполнять эту программу последовательно, по частям. В любой момент можно остановиться и изучить полученные результаты
* Блокнот позволяет только редактировать код и отображать выполнения. Все вычисления производятся ядром (kernel) - интерпретатором Python, запущенным на вашем компьютере
* При первом открытии блокнота, или после перезапуска ядра все нужно вычислять заново
* Блокнот сохраняет содержимое ячеек вывода, поэтому если нужно только посмотреть на результаты, то повторные вычисления не требуются

### Порядок вычислений в блокноте

Код в блокноте записан последовательно. Однако вы можете изменить порядок выполнения, запуская отдельные ячейки в произвольном порядке.
При этом значения переменных и другие результаты вычислений будут определяться порядком **выполнения** (т.е. номером в квадратных скобках после In [1] ) ячеек с кодом, а не порядком их **записи** в блокноте!

При перезапуске ядра все результаты в памяти стираются. Их нужно будет получать заново, запуская соответствующие ячейки.

<img src="Image_introduction/Jupyter_struct.png" alt="Юпитер-структура" width="900">

Ячейки бывают в двух режимах:
- режим редактирования;
- командный режим.

В режиме редактирования цвет рамки ячейки - зеленый. В нем вы можете редактировать текст внутри ячейки (довольно неожиданно - не правда ли)

<img src="Image_introduction/Jupyter_green_ceil.png" alt="Юпитер-режим редактирования" width="900">

В командном режиме цвет рамки ячейки - синий. В нем вы можете перемещаться между ячейками, выполнять над ними операции - копировать, вырезать, вставлять новые, удалять и т.д.

<img src="Image_introduction/Jupyter_blue_ceil.png" alt="Юпитер-командный режим" width="900">

#### Как перейти из одного режима в другой? 
Из командного в редактирование (из синего в зеленый) - жмем либо курсором мыши на нужной ячейке, либо же выбираем ее стрелочками и жмем *Enter*

Из режима редактирования в командный режим (из зеленого в синий) - нажимаем *Esc*

### Выполнение ячейки 

Выполнить ячейку можно двумя способами:
- Ctrl + Enter - выполняет ячейку, фокус остается на той же ячейке, режим становится командным (синим);
- Shift + Enter - выполняет ячейку, создает за ней новую пустую ячейку и переключает на нее фокус.

### Типы ячеек

Итак, пришло наконец время задаться вопросом, как же написан текст в данном блокноте? В блокноте ячейки могут быть трех видов:
- для кода (Code);
- для размеченного текста (Markdown);
- для неразмеченного (сырого) текста (Raw).

Изменить тип ячейки можно в следующем списке:
<img src="Image_introduction/Jupyter_type_ceil.png" alt="Юпитер-тип ячейки" width="900">

Отлично, вы увидели, где можно изменить тип ячейки, а теперь **забудьте** это! В юпитере, как и везде есть горячие клавиши, с помощью которых вы можете облегчить себе жизнь. Я не хочу заставлять вас учить все горячие клавиши, но знать пару основных вы просто обязаны!

<div class="alert alert-block alert-danger">
<b>АЛАРМ:</b> Все нижеперечисленные горячие клавиши работают только в командном режиме (ячейка должна быть СИНЯЯ). Однако, есть и горячие клавиши, которые работают в режиме редактирования.
</div>

### Горячие клавиши перехода между типами:
- y - переход в режим кода (Code);
- m - переход в режим размеченного текста (Markdown);
- r - переход в режим сырого текста (Raw).

По факту режим сырого текста (Raw) практически никогда не используется, поэтому запомните лишь клавиши m (markdown) и y (~~yode~~ code)

Раз уж мы заговорили о горячих клавишах, то перечислю те, которые по-моему мнению обязаны знать все, кто работает с ноутбуками больше, чем 30 секунд:
- a (above) - добавить ячейку сверху;
- b (below) - добавить ячейку ниже;
- x (даже ассоциацию не буду писать, тут как в ворде) - вырезать ячейку;
- c (аналогично...) - скопировать ячейку;
- v (ну вы поняли уже...) - вставить ячейку.

Чтобы открыть полный список горящих клавиш нажмите - Help/Keyboard ShortCuts

### А что если мой ноутбук зависнет?

Как понять, что ваш ноутбук работает, а команда не выполнилась? Ячейка выполняется, когда в квадратных скобках вместо цифры висит знак зведочки - *. При работающем ноутбуке в правом верхнем углу возле надписи Python 3 вместо окружности появляется залитый серым цветом круг.

<img src="Image_introduction/Zavis.png" alt="Зависла ячейка" width="900">

Остановить такой ноутбук можно при нажатии квадрата на панеле команд (справа от Run), либо в Kernel/Interrupt

<img src="Image_introduction/Stop.png" alt="Меню Кернел" width="600">

В меню Kernel есть также возможность перезапустить ядро (все операции сбросятся и их нужно будет выполнять заново, однако все выходы продолжат отображаться), перезагрузить ядро и очистить выходы, а также перезапустить ядро и запустить все заново.
